```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
df = pd.read_csv("data_Training.csv")
X = df.iloc[:, :-1]
y_raw = df.iloc[:, -1]
```

```python
le = LabelEncoder()
y = le.fit_transform(y_raw)
```

```python
model = DecisionTreeClassifier()  
model.fit(X, y)
```

```python
plt.figure(figsize=(30, 40)) 


plot_tree(
    model,
    feature_names=X.columns,
    class_names=le.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)

plt.title("Decision Tree for Disease Prediction", fontsize=20)
plt.show()
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X)

accuracy = accuracy_score(y, y_pred)

print(f"\n Decision Tree Accuracy on Training Data: {accuracy * 100:.2f}%")
```

```python

symptoms = list(X.columns)

print(" Welcome to Symptom Checker!")
print("Please answer 1 for YES or 0 for NO for the following symptoms:\n")

user_input = []
for symptom in symptoms:
    while True:
        val = input(f"Do you have '{symptom}'? (1 = Yes, 0 = No): ")
        if val in ['0', '1']:
            user_input.append(int(val))
            break
        else:
            print("Please enter only 0 or 1.")

user_input_array = np.array([user_input])  # 2D
prediction = model.predict(user_input_array)
predicted_disease = le.inverse_transform(prediction)[0]

print("\n Based on your symptoms, the predicted disease is:")
print(f" {predicted_disease}")
```
